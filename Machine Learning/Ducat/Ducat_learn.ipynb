##  Creating the connection with the webpage

```python
import requests as req
```

```python
response=req.get("https://www.mohfw.gov.in/")
webdata=response.text  #return all the data of the web page
#print(webdata)
```

## Table of the Corona Virus

```python
my_list=pd.read_html(webdata)
#print(my_list)
print(len(my_list))
```

```python
corona=my_list[0]
corona
```

```python
corona.drop('S. No.',axis=1,inplace=True)
corona
```

```python
corona.columns=['State','Active_Cases','Discharged','Deaths','Total_Cases']
```

```python
corona
```

```python
corona.drop([35,36,37,38,39,40],axis=0,inplace=True)
corona
```

```python
corona.dtypes
```

```python
corona.replace({'0#':'0'},inplace=True)
```

## Evaluation on the dataset 

### Active Cases Number

```python
corona.Active_Cases=corona.Active_Cases.map(int)
corona.Active_Cases.sum()
```

### Total Cases Number

```python
corona.Total_Cases=corona.Total_Cases.map(int)
corona.Total_Cases.sum()
```

### Total Deaths Number

```python
corona.Deaths=corona.Deaths.map(int)
corona.Deaths.sum()
```

### Total Recovered Cases

```python
corona.Discharged=corona.Discharged.map(int)
corona.Discharged.sum()
```

## Death Ratio

```python
def death_ratio(deaths,total_cases):
    return (deaths/total_cases*100)
```

```python
corona['Death_ratio']=death_ratio(corona.Deaths,corona.Total_Cases)
```

```python
corona[corona.Death_ratio==corona.Death_ratio.max()]
```

```python
corona.sort_values(by='Death_ratio',ascending=False)
```
