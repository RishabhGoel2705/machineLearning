```python
import requests
```

```python
resp=requests.get("https://www.mohfw.gov.in/") #it will create coneection with webpage
webdata=resp.text #returns all data of webpage
#print(webdata)
```

```python
df_list=pd.read_html(webdata)
```

```python
len(df_list)
```

```python
corona=df_list[0]
```

```python
corona.drop('S. No.',axis=1,inplace=True)
```

```python
corona
```

```python
corona.drop([34,35,36,37,38,39],axis=0,inplace=True)
```

```python
corona
```

```python
corona.columns=['State','Confirn_cases','Discharged','Deaths']
```

```python
corona
```

```python
corona.dtypes
```

```python
corona.replace({'0#':'0'},inplace=True)
```

```python
corona.Confirn_cases=corona.Confirn_cases.map(int)
corona.Discharged=corona.Discharged.map(int)
corona.Deaths=corona.Deaths.map(int)
```

```python
corona.dtypes
```

```python
corona.Confirn_cases.sum()
```

```python
corona.Deaths.sum()
```

```python
corona.sort_values(by='Confirn_cases',ascending=False)
```

```python
def deathratio(death,confirm_cases):
    return (death/confirm_cases)*100
```

```python
corona['Death_Ratio']=deathratio(corona.Deaths,corona.Confirn_cases)
```

```python
corona.Death_Ratio.max()
```

```python
corona[corona.Death_Ratio==corona.Death_Ratio.max()]
```

```python
corona.Death_Ratio.mean()
```

```python
top10=corona.sort_values(by='Confirn_cases',ascending=False).head(10)
```

```python
import matplotlib.pyplot as plt
import seaborn as sb
```

```python
plt.figure(figsize=(20,7))
sb.barplot(x='State',y='Confirn_cases',data=top10)
plt.savefig('top10.png')
```
